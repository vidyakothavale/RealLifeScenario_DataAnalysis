```python

expresult['Weight slab as per courier company(kg)'] = expresult['Total Weight as per courier company(kg)'].apply(lambda x : math.ceil(x*2)/2)
expresult
```

```python
expresult = pd.merge(expresult, 
                      couriercomp_invoice[['Order ID','Zone']], 
                      on ='Order ID')
expresult
```

```python

expresult.rename(columns={'Zone':'Delivery zone as per X'},inplace=True)
expresult.head()
```

```python

# Delivery zone charge by courier company

expresult['Delivery zone charge by courier company'] = expresult['Delivery zone as per X']
expresult.head()
```

```python
#Calcuation of 'Expected charges as per X'

# creating the dataframe having 'Order ID','Weight slab as per X(kg)','Delivery zone as per X'

calculation =pd.DataFrame(expresult[['Order ID','Weight slab as per X (KG)','Delivery zone as per X']])
calculation
```

```python
calculation = pd.merge(calculation, 
                      couriercomp_invoice[['Order ID','Type of Shipment']], 
                      on ='Order ID')

calculation.head()
```

```python
couriercomp_Rate
```

```python
calculation['Expected charge as per X(Rs)']=0
```

```python
calculation
```

```python
for i in range(0,123):
    
    if ((calculation['Delivery zone as per X'][i]=='d') and (calculation['Type of Shipment'][i]=='Forward charges')):
        calculation['Expected charge as per X(Rs)'][i] = couriercomp_Rate['fwd_d_fixed'] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['fwd_d_additional'][0]
        
    elif ((calculation['Delivery zone as per X'][i]=='d') and (calculation['Type of Shipment'][i]=='Forward and RTO charges')):
        calculation['Expected charge as per X(Rs)'][i] = couriercomp_Rate['fwd_d_fixed'][0] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['fwd_d_additional'][0] + couriercomp_Rate['rto_d_fixed'][0] 
        
    elif ((calculation['Delivery zone as per X'][i]=='b') and (calculation['Type of Shipment'][i]=='Forward charges')):
        calculation['Expected charge as per X(Rs)'][i] = couriercomp_Rate['fwd_b_fixed'][0] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['fwd_b_additional'][0]
        
    elif ((calculation['Delivery zone as per X'][i]=='b') and (calculation['Type of Shipment'][i]=='Forward and RTO charges')):
        calculation['Expected charge as per X(Rs)'][i] = couriercomp_Rate['fwd_b_fixed'][0] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['fwd_d_additional'][0] + couriercomp_Rate['rto_b_fixed'][0] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['rto_b_additional'][0]
        
    elif ((calculation['Delivery zone as per X'][i]=='e') and (calculation['Type of Shipment'][i]=='Forward charges')):
        calculation['Expected charge as per X(Rs)'][i] = couriercomp_Rate['fwd_e_fixed'][0] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['fwd_e_additional'][0]
        
    elif ((calculation['Delivery zone as per X'][i]=='e') and (calculation['Type of Shipment'][i]=='Forward and RTO charges')):
        calculation['Expected charge as per X(Rs)'][i] = couriercomp_Rate['fwd_e_fixed'][0] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['fwd_e_additional'][0] + couriercomp_Rate['rto_e_fixed'][0] + ((calculation['Weight slab as per X (KG)'][i]/0.5)-1)*couriercomp_Rate['rto_e_additional'][0]
        
    calculation['Expected charge as per X(Rs)'].append(calculation['Expected charge as per X(Rs)'])

```

```python
expresult = pd.merge(expresult, 
                      calculation[['Order ID','Expected charge as per X(Rs)']], 
                      on ='Order ID')

expresult.head()
```

```python
expresult = pd.merge(expresult, 
                      couriercomp_invoice[['Order ID','Billing Amount (Rs.)']], 
                      on ='Order ID')
expresult
```

```python

# Renaming the column

expresult.rename(columns={'Billing Amount (Rs.)':'Charged billed by courier company(Rs)'},inplace=True)
expresult.head()
```

```python
expresult['Difference between expected charges and billed charges(Rs)'] = round(expresult['Expected charge as per X(Rs)']-expresult['Charged billed by courier company(Rs)'],3)
expresult
```

```python
expresult.to_excel("ExpResults1.xlsx")
```

```python
count=expresult[expresult['Difference between expected charges and billed charges(Rs)']==0].count()['Difference between expected charges and billed charges(Rs)']
count
```

```python
# Count of Total Orders - Over Charged

courier=expresult[expresult['Difference between expected charges and billed charges(Rs)']<0].count()['Difference between expected charges and billed charges(Rs)']
courier
```

```python
# Count of Total Orders - Under Charged

countoftotalor=expresult[expresult['Difference between expected charges and billed charges(Rs)']>0].count()['Difference between expected charges and billed charges(Rs)']
countoftotalor
```

```python
# sum of Total Orders - Correctly Charged

sumofcorrect=round(expresult[expresult['Difference between expected charges and billed charges(Rs)']==0].sum()['Expected charge as per X(Rs)'],2)
sumofcorrect
```

```python

# sum of Total Orders - Over Charged

sumoforders=round(expresult[expresult['Difference between expected charges and billed charges(Rs)']<0].sum()['Expected charge as per X(Rs)'],2)
sumoforders
```

```python
# sum of Total Orders - Under Charged

sumofordersuc=round(expresult[expresult['Difference between expected charges and billed charges(Rs)']>0].sum()['Expected charge as per X(Rs)'],2)
sumofordersuc
```

```python

data = {'Count': [count,courier,countoftotalor],
        'Sum': [sumofcorrect,sumoforders,sumofordersuc]}
```

```python
   

Summary = pd.DataFrame(data, index=['Total Orders - Correctly Charged','Total Orders - Over Charged','Total Orders - Under Charged'])
Summary
```
